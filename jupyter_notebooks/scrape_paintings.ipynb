```python
import time
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from urllib.request import urlretrieve
import os
import re
```

```python
DOWNLOAD_DIR = 'data/paintings'
```

```python
def scroll_down():
    time.sleep(5)
    # define initial page height for 'while' loop
    last_height = driver.execute_script("return document.body.scrollHeight")
    count = 0
    while True:
        count += 1
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        try:
            load_xpath = "//span[@class='load-more-text']"
            loads = driver.find_elements_by_xpath(load_xpath)
            mouse_click(loads[0])

        except:
            print('no See More button found...')

        time.sleep(5)
        new_height = driver.execute_script("return document.body.scrollHeight")

        print(count, new_height)

        if new_height == last_height:
            break
        else:
            last_height = new_height

def save_img_src(url, file_no, sleep_time=0.25):
    file_name_full = '%s/painting_%s.%s' % (DOWNLOAD_DIR, file_no, 'JPEG')
    try:
        urlretrieve(url, file_name_full)
        if not file_no % 100:
            print('wrote from url %s' % file_name_full)
    except IOError as e:
        print('Bad URL?', e)
    time.sleep(sleep_time)
```

```python
chromedriver = "/Users/PandaGongfu/Downloads/chromedriver_2"
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(chromedriver)

target_url_prefix = 'https://www.flickr.com/groups/painting/pool/page'
start_no = 5243

error_pages = []
```

```python
for page in range(505, 570):
    try:
        print('processig page {:4d}'.format(page))
        driver.get(target_url_prefix+str(page))

        # scroll down to the bottom of the page
        scroll_down()

        image_xpath = "//div[@class='view photo-list-photo-view requiredToShowOnServer awake']"
        imgs_1 = driver.find_elements_by_xpath(image_xpath)

        image_xpath = "//div[@class='view photo-list-photo-view awake']"
        imgs_2 = driver.find_elements_by_xpath(image_xpath)

        imgs = imgs_1+imgs_2
        urls = []
        for img in imgs:
            urls.append(''.join(['https:', re.findall(r'url\(\"(.+)\"', img.get_attribute('style'))[0]]))

        print(len(urls))    
        for file_no, url in enumerate(urls):
            save_img_src(url, start_no)
            start_no += 1
    except:
        error_pages.append(page)
        
```
